```python
#Use get_dummies method from panda class to create dummy variable and store those
dummies = pd.get_dummies(dataset.Species)
#Display last 5 dummies
dummies.tail()
```

```python
#Concate the newly created dummy variables with the loaded data and store in merg
merged_data = pd.concat([dataset,dummies], axis=1)
```

```python
merged_data
```

```python
#Display merged_data
merged_data.head()
```

```python
#Drop the Species column since we've done one hot encoding, hence Species column
final_data = merged_data.drop(columns='Species')
```
